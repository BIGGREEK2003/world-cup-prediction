```python

team_stats['xg_total'] = team_stats['home_xg'] + team_stats['away_xg']
team_stats['xg_diff'] = team_stats['home_xg'] - team_stats['away_xg']
team_stats['clearance_diff'] = team_stats['home_clearances'] - team_stats['away_clearances']
team_stats['long_ball_diff'] = team_stats['home_long_balls'] - team_stats['away_long_balls']

team_features = team_stats.groupby('home_team').agg({
    'xg_total': 'mean',
    'xg_diff': 'mean',
    'clearance_diff': 'mean',
    'long_ball_diff': 'mean',
    'attendance': 'mean'
}).reset_index().rename(columns={'home_team': 'team'})
```

```python

players_df['age'] = 2022 - players_df['birth_year']
player_features = players_df.groupby('team').agg({
    'age': 'mean',
    'minutes_90s': 'sum'
}).reset_index().rename(columns={'age': 'avg_age', 'minutes_90s': 'total_minutes'})
```

```python

team_profiles = team_features.merge(player_features, on='team', how='inner')
```

```python
world_cups['winner_flag'] = 1
team_profiles['winner_flag'] = team_profiles['team'].apply(
    lambda x: 1 if x in world_cups['winner'].values else 0
)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

X = team_profiles.drop(columns=['team', 'winner_flag'])
y = team_profiles['winner_flag']

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
winner_probs = model.predict_proba(X)[:, 1]
team_profiles['win_probability'] = winner_probs
top_teams = team_profiles.sort_values('win_probability', ascending=False)
print(top_teams[['team', 'win_probability']].head(10))
```

```python

team_profiles_sorted = team_profiles.sort_values('win_probability', ascending=False)
predicted_winner = team_profiles_sorted.iloc[0]
print(f"The predicted World Cup winner is: {predicted_winner['team']}")
print(f"Win probability: {predicted_winner['win_probability']:.2%}")
```

```python
top_value = team_profiles['win_probability'].max()
top_teams = team_profiles[team_profiles['win_probability'] == top_value]
if len(top_teams) == 1:
    winner = top_teams.iloc[0]['team']
else:
    winner = top_teams.sort_values('xg_total', ascending=False).iloc[0]['team']
print(f"The predicted World Cup winner is: {winner}")
```
